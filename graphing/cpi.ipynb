```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
df = pd.read_csv('CPIAUCSL.csv',sep=',',header=0,names=['DATE','CPIAUCSL'],parse_dates=[1],index_col='DATE',encoding = "ISO-8859-1")
```

Cast data to floats.

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
ax = df.plot(title='Consumer Price Index')
ax.legend(['Consumer Price Index'])
```
