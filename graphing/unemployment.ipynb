Import helper libraries for analysis and plotting. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Do some basic configuration for plots.

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)
```

Read CSV

```python
df = pd.read_csv('UNRATE.csv',sep=',',header=0,names=['Date','Unemployment Rate'],parse_dates=[1],index_col='Date',encoding = "ISO-8859-1")
```

Normalize the data by renaming the column `UNRATE` to the meaningful `Unemployment Rate` and cast the column to all floats.

```python
# df = df.rename(columns={'UNRATE': 'Unemployment Rate'})
df['Unemployment Rate'] = pd.to_numeric(df['Unemployment Rate'], errors='coerce')
```

Plot the data.

```python
df.plot(title='Unemployment Rate by Year')
```
